# 37 Truss Bar

```python
from pysead import Truss_2D
```

```python
nodes = []

for i in range(11):
    for j in range(2):
        nodes.append([i,j])

del nodes[1]
del nodes[-1]
nodes = {key+1: nodes[key] for key in range(20)}

elements=  {1:[1, 3], 2:[2, 3], 3:[4, 3], 4:[3, 5], 5:[4, 5], 6:[6, 5], 7:[5, 7], 8:[6, 7], 9:[8, 7], 10:[7, 9], 
         11:[8, 9], 12:[10, 9], 13:[9, 11], 14:[10, 11], 15:[10, 13], 16:[11, 13], 17:[12, 13], 18:[12, 15], 19:[13, 15], 20:[14, 15], 
         21:[14, 17], 22:[15, 17], 23:[16, 17], 24:[16, 19], 25:[17, 19], 26:[18, 19], 27:[19, 20], 28:[1, 2], 29:[2, 4], 30:[4, 6], 
         31:[6, 8], 32:[8, 10], 33:[10, 12], 34:[12, 14], 35:[14, 16], 36:[16, 18], 37:[18, 20]}

elasticity = {key: 2.11e6 for key in elements}
cross_area = {key: 0.004 for key in elements}
supports = {1:[1,1], 20:[1,1]}
forces = {key: [0,-98.1] for key in range(2,19,2)}
```

```python
Truss = Truss_2D(nodes = nodes,
                 elements = elements,
                 supports = supports,
                 forces = forces,
                 elasticity = elasticity,
                 cross_area = cross_area)
```

```python
Truss.Draw_Truss_Setup(figure_size=[20,5], length_of_arrow=0.3, width_of_arrow=0.05, offset = 0.015)
```

```python
Truss.Solve()
```

```python
Truss.Draw_Truss_Axial_Force_Map(figure_size=[20,5], color_bar_orientation='horizontal')
```

```python
Truss.member_forces_
```

```python
Truss.Draw_Truss_Displacements(figure_size=[20,5], offset = 0.015, magnification_factor=0.1)
```
