```python
from pysead.Frame_2D import Frame_2D
from pysead.Frame_2D import Member_2D
```

```python
M1 = Member_2D(1,14.7, 29_000, 800, {1:[0,0], 2:[0,20*12]}, [0,1])
M2 = Member_2D(2,14.7, 29_000, 800, {2:[0,20*12], 3:[20*12,20*12]}, [1,0])
M3 = Member_2D(3,14.7, 29_000, 800, {3:[20*12,20*12], 4:[20*12,0]}, [0,0])
M1.Add_Load_Full_Uniform(0.1)
M2.Add_Load_Point(75,10*12)
```

```python
M1.Plot_Diagrams()
```

```python
M2.Plot_Diagrams()
```

```python
M3.Plot_Diagrams()
```

```python
Frame = Frame_2D()
Frame.Compile_Frame_Member_Properties([M1,M2,M3])
Frame.Add_Load_Node({2:[25,0,0]})
Frame.supports = {1:[1,1,1], 4:[1,1,0]}
```

```python
Frame.Draw_Frame_Setup(figure_size=[10,10], length_of_arrow=30, linewidth = 4)
```

```python
Frame.Solve()
```
