```python
from pysead.Frame_2D import Frame_2D
from pysead.Frame_2D import Member_2D
```

```python
M1 = Member_2D(member_number = 1, 
               nodes = {1:[0,0], 2:[0,3]},
               area = 2e-2,
               inertia = 5e-5,
               elasticity= 210_000_000)
M2 = Member_2D(member_number = 2, 
               nodes = {2:[0,3], 3:[4,3]},
               area = 2e-2,
               inertia = 5e-5,
               elasticity= 210_000_000)
M3 = Member_2D(member_number = 3, 
               nodes = {3:[4,3], 4:[4,0]},
               area = 2e-2,
               inertia = 5e-5,
               elasticity= 210_000_000)
Frame = Frame_2D()
Frame.Compile_Frame_Member_Properties([M1,M2,M3])
Frame.supports = {1:[1,1,1], 4:[1,1,1]}
Frame.Add_Load_Node({2:[-20,0,0], 3:[0,0,12]})
Frame.Solve()
```

```python
Frame.Draw_Frame_Setup(figure_size=[10,10], length_of_arrow = 0.5, arrow_line_width=2,radius_of_arc=0.5,linewidth=2)
```
