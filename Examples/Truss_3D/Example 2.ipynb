```python
from pysead import Truss_3D
%matplotlib widget
```

```python
nodes = {1: [-6*12,8*12,0], 2:[12*12,8*12,0], 3:[6*12,-8*12,0], 4:[-12*12,-8*12,0], 5:[0,0,24*12]}
elements = {1:[5,1], 2:[5,2], 3:[5,3], 4:[4,5]}
elasticity = {key: 10_000 for key in elements}
cross_area = {key: 8.4 for key in elements}
supports = {1:[1,1,1], 2:[1,1,1], 3:[1,1,1], 4:[1,1,1]}
forces = {5:[0,100,-50]}
```

```python
Four_Bar_Truss = Truss_3D(nodes = nodes,
                          elements=elements,
                          elasticity=elasticity,
                          cross_area=cross_area,
                          supports=supports,
                          forces= forces)
```

```python
Four_Bar_Truss.Solve()
```

```python
Four_Bar_Truss.Draw_Truss_Setup(length_of_arrow=50)
```
